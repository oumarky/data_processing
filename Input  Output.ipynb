# 4 - Exporter un fichier Excel
* installer la librairie xlrd et openpyxl
* methode save()

```python
### importons un jeux de données quelconque
url = "https://www.data.gov.bf/sites/default/files/listedescandidatsauxelectionslegislativeslisteprovincialeanptic.xlsx"
list_candidats = pd.read_excel(url)
list_candidats
```

```python
### nous alons etraire et exporter les candidats en fonction de leur sexe (basé sur 'SEXE_TITULAIRE')
men = list_candidats[list_candidats['SEXE_TITULAIRE']== "M"]
women = list_candidats[list_candidats['SEXE_TITULAIRE']=="F"]
women
```

```python
## Créer/initialiser d'abord un ExcelWritter Object
excel_file = pd.ExcelWriter("candidates.xlsx")
# identique a un objetc pd => pd.DataFrame()
# utiliser ensuite to_excel()
men.to_excel(excel_file,sheet_name='men',index = False)
```

```python
# same for women
women.to_excel(excel_file,sheet_name='women',index = False)
```

```python
### Nous pouvons aussi specifier les colonnes à enregistrer
# same for women
women.to_excel(excel_file,sheet_name='sub_women',index = False, columns= ['REGION_NOM',	'PROVINCE_NOM',	'CIRCONSCRIPTION',	'SIGLE'])
```

```python
### utiliser maintenant la methode save() pour enregistrer le fichier excel_file que nous nous avons initialisé
excel_file.save()
```

## 5- Importer et Exporter un fichier Json

```python
## commencer par importer la lib json
import json
```

#### La fonction load()

```python
def read_json(json_file):
    with open('filename.json') as f:
        data = json.load(f)
    return data
```

#### La fonction dump()

```python
# Prend en parametre un dictionnaire
def save_to_json(dic):
    with open("filename.jsonl", encoding='utf8') as outf: # utiliser utf8 pour assurer le bon formatqge des données
                                                            # textuelles
        json.dump(dic, ensure_ascii=False)
        outf.write()
```

## 6- Importer et Exporter un fichier Texte

#### Lire un document entier

```python
# fxn1 pour lire des données dans un fichier txt
def read_txt_file(file):
    with open(file) as f:
        body = f.read()
    return body# vous pouvez utilise .splitlines() pour obtenir la liste de lignes
```

#### Lire un document ligne par ligne

```python
# fxn2 pour lire des données dans un fichier txt
def read_txt_file_line_by_line(file):
    with open(file) as f:
        lines = f.readlines()# # liste contenant toutes les lignes
        """vous pouvez utiliser une bouble pour recuperer chaque ligne
            ou utiliser directement la focntion readline()
        """
    return lines
```

#### Exporter des données dans un fichier txt
* https://www.pythontutorial.net/python-basics/python-write-text-file/

```python
# fxn pour ecrire des données dans un fichier txt
def write_to_txt(data):
    with open('filename.txt', 'w') as f: # 'w' ==> ouvrir un nouveau fichier en mode écriture                                         # 
                                        # 'a' ==> ouvrir un nouveau fichier en mode ajout, un nouveau fichier sera crée, s'il n'existe pas encore 
        f.write(data) # ecriture des données
        
### Quelques notes
# f.write('\n') => permet d'ecrire un String dans un txt
# f.writelines('\n') => permet d'ecrire une liste de string dans un fichier txt.
# '\n' => signifie retour à la ligne après ecriture
######
```
